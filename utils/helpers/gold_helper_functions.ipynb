```python
from pyspark.sql.functions import *
from pyspark.sql import DataFrame
```

```python
def read_df(table_name: str) -> DataFrame:
    """
    Reads the table from the database and returns the Spark DataFrame

    args:
        table_name: name of the table
    """
    return spark.read.table(table_name)
```

```python
def save_table(df: DataFrame, table_name: str) -> None:
    """
    Saves the DataFrame to the silver table name provided as input

    args:
        df: Spark DataFrame
        table_name: name of the table
    """
    df.write.mode("overwrite").saveAsTable(f"PROJECT1001.GOLD.{table_name}_gold")
```

```python
def total_items_per_order(df: DataFrame) -> DataFrame:
    """
    Get total items purchased per order

    args:
        df: Orders_details DataFrame
    """
    return df.groupBy("order_id").agg(count("item_id").alias("total_items"))
```

```python
def total_orders_per_day(df: DataFrame) -> DataFrame:
    """
    Get total orders placed per day

    args:
        df: Orders DataFrame
    """
    return df.groupBy("order_date").agg(count("order_id").alias("total_orders_per_day"))
```

```python
def returning_customers_per_item(df: DataFrame) -> DataFrame:
    """
    Get the number of returning customers per item

    args:
        df: Orders_details DataFrame
    """
    return df.groupBy("item_id", "customer_id").agg(count("customer_id").alias("returning_customers_per_item")).filter("returning_customers_per_item > 1")
```

```python
def customer_spending(df: DataFrame) -> DataFrame:
    """
    Get the total spending of each customer

    args:
        df: Orders DataFrame
    """
    return df.groupBy("customer_id","name").agg(sum("total_amount").alias("total_spending"))
```

```python
def active_customers_per_state(df: DataFrame) -> DataFrame:
    """
    Get the number of active customers per state

    args:
        df: Customers DataFrame
    """
    return df.filter(df.is_active).groupBy("state").agg(count("customer_id").alias("active_customers_per_state"))
```

```python
def number_of_payments_per_day_by_method(df: DataFrame) -> DataFrame:
    """
    Get the number of payments per day for each payment method

    args:
        df: Payments DataFrame
    """
    return df.groupBy("payment_date").pivot("payment_method", ["paypal", "upi", "debit_card", "credit_card"]).agg(count("payment_id").alias("number_of_payments_per_day_by_method")).fillna(0)

def shippings_per_day_by_method(df: DataFrame) -> DataFrame:
    """
    Get the number of shippings per day for each shipping method

    args:
        df: Shipments DataFrame
    """
    return df.groupBy("shipping_date").pivot("shipping_method", ["standard", "express"]).agg(count("shipping_id").alias("shippings_per_day_by_method")).fillna(0)
```

```python
def orders_per_city(df: DataFrame) -> DataFrame:
    """
    Get the number of orders per city
    args:
        df: Orders_details DataFrame
    """
    return df.groupBy("city").agg(count("order_id").alias("order_per_city"))

def orders_per_state(df: DataFrame) -> DataFrame:
    """
    Get the number of orders per state
    args:
        df: Orders_details DataFrame
    """
    return df.groupBy("state").agg(count("order_id").alias("order_per_state"))

def orders_per_country(df: DataFrame) -> DataFrame:
    """
    Get the number of orders per country
    args:
        df: Orders_details DataFrame
    """
    return df.groupBy("country").agg(count("order_id").alias("order_per_country"))
```

```python
def revenue_per_category(df: DataFrame) -> DataFrame:
    """
    Get the total revenue per category
    args:
        df: Orders_details DataFrame
    """
    return df.groupBy("category").agg(sum("total_amount").alias("revenue_per_category"))
```
