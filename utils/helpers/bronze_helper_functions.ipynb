```python
from pyspark.sql.types import StructType
```

```python
def read_data(path: str, schema: StructType, header: bool, table_name: str) -> None:

    '''
    Reads raw data from give path and saves it as a table in the bronze database

    args:
        path: path to the raw data
        schema: schema of the data
        header: whether the data has a header
        table_name: name of the table
    '''

    from pyspark.sql.functions import current_timestamp, col

    print(f"\n\nReading data for {table_name} from {path}")
    df = spark.read.schema(schema).csv(path, header=header)
    print(f"Completed reading data {table_name}")

    # add ingestion metadata
    df=df.withColumn("ingestion_timestamp", current_timestamp())\
                        .withColumn("file_name", col("_metadata.file_path"))
    print(f"Added ingestion metadata for {table_name}")

    print(f"Saving data for {table_name} as bronze table")
    #Save the data as bronze table
    df.write.mode("overwrite").saveAsTable(f"PROJECT1001.BRONZE.{table_name}_bronze")
    print(f"Completed saving data for {table_name} as bronze table")
```
