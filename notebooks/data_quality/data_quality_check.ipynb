- Row counts for each table : Confirm ingestion worked.
- Schema correctness : Confirm columns exist and types match expectations.
- Presence of ingestion metadata columns : Verify ingestion metadata was added.
- Data types are correct
- No unexpected nulls in primary identifiers : Primary keys should not be null.
- Duplicate Checks
- If referential integrities are maintained

This prevents debugging problems later in the Silver layer.

```sql
%sql
--Check row count in the tables with primary key as null

SELECT "CUSTOMERS_BRONZE" TABLE_NAME, COUNT(*) FROM CUSTOMERS_BRONZE WHERE customer_id IS NULL
UNION
SELECT "ITEMS_BRONZE" TABLE_NAME, COUNT(*) FROM ITEMS_BRONZE WHERE item_id IS NULL
UNION
SELECT "ORDERS_BRONZE" TABLE_NAME, COUNT(*) FROM ORDERS_BRONZE WHERE order_id IS NULL
UNION
SELECT "PAYMENTS_BRONZE" TABLE_NAME, COUNT(*) FROM PAYMENTS_BRONZE WHERE payment_id IS NULL
UNION
SELECT "SHIPPINGS_BRONZE" TABLE_NAME, COUNT(*) FROM SHIPPINGS_BRONZE WHERE shipping_id IS NULL;
```

```sql
%sql
--Check row count in the tables with primary key as null

SELECT "CUSTOMERS_BRONZE" TABLE_NAME, customer_id, COUNT(*) FROM CUSTOMERS_BRONZE GROUP BY customer_id HAVING COUNT(*)>1
UNION
SELECT "ITEMS_BRONZE" TABLE_NAME, item_id, COUNT(*) FROM ITEMS_BRONZE GROUP BY item_id HAVING COUNT(*)>1
UNION
SELECT "ORDERS_BRONZE" TABLE_NAME, order_id, COUNT(*) FROM ORDERS_BRONZE GROUP BY order_id HAVING COUNT(*)>1
UNION
SELECT "PAYMENTS_BRONZE" TABLE_NAME, payment_id, COUNT(*) FROM PAYMENTS_BRONZE GROUP BY payment_id HAVING COUNT(*)>1
UNION
SELECT "SHIPPINGS_BRONZE" TABLE_NAME, shipping_id, COUNT(*) FROM SHIPPINGS_BRONZE GROUP BY shipping_id HAVING COUNT(*)>1;
```
